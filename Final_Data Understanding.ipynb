```python
# Aggregate df_eng to customer level

sub_set = set(df_eng['subid'].tolist())
```

```python
# check duplicate for each customer to see whether engagement record could happen more than once per day for an individual customer

df_eng.duplicated(subset = ['subid', 'date']).sum()

# The result tells that there is no duplicate engagement record for each individual
```

```python
# Daily Average

df_eng_daily = df_eng.groupby(by = ['subid', 'payment_period']).mean()

col = list(df_eng_daily.columns)

for i in range(len(col)):
    col[i] = 'daily_' + col[i]
    
df_eng_daily.columns = col

df_eng_daily = df_eng_daily.unstack(level = 1)

col2 = []
for i in col:
    for j in range(4):
        col2.append(i + '_period_' + str(j))
        
df_eng_daily.columns = col2

df_eng_daily
```
